```python
import matplotlib.pyplot as plt #pyplot easy to make vishulization of graphs,charts
import numpy as np

x = np.array([2021,2022,2023,2024,2025,2026])#changing list to numpy array is making fast and much easy
y = np.array([0,10000,60000,100000,200000,300000])
y1 =np.array([0,1000,2500,5000,10000,25000])
```

```python
plotProps = dict( marker = "o",#marking the plots in given sign
         markersize = 10,
         markerfacecolor = "yellow",
         markeredgecolor = "red",
         markeredgewidth = 3,
         linestyle = "solid",#dashed,dashdot,solid,none
         linewidth = 2,
         )
```

```python
plt.plot(x,y,color = "grey",**plotProps)
plt.plot(x,y1,color = "Blue",**plotProps)
plt.title("Subscriber count",
          fontsize = 15,
          fontweight = "bold",
          color = "blue",
         family = "Arial" )
plt.xlabel("year",
           fontsize = 20,
           family = "arial")
plt.ylabel("subscriber count",
           fontsize = 20,
           family = "arial")
plt.subplots_adjust(left = 0.15,bottom = 0.2)
#plt.xticks(x)   which means it'll remove the unwanted x ticks
#plt.tick_params(axis = "x",     here we can use y or even the word "both"
#                labelcolor = "red",
#               labelsize = 10) which means here we can change the x or y axis's styles,fonts,color,customizing seperatly. 
plt.grid(axis = "both",
         color = "brown",
         linestyle = "dashdot",
         linewidth = 2)
plt.show()
```
