```python
import matplotlib.pyplot as plt #pyplot easy to make vishulization of graphs,charts
import numpy as np

x = np.array([2021,2022,2023,2024,2025,2026])#changing list to numpy array is making fast and much easy
y = np.array([0,10000,60000,100000,200000,300000])
y1 =np.array([0,1000,2500,5000,10000,25000])
```

```python
plotProps = dict( marker = "o",#marking the plots in given sign
         markersize = 10,
         markerfacecolor = "yellow",
         markeredgecolor = "red",
         markeredgewidth = 3,
         linestyle = "solid",#dashed,dashdot,solid,none
         linewidth = 2,
         )
```

```python
plt.plot(x,y,color = "grey",**plotProps)
plt.plot(x,y1,color = "Blue",**plotProps)
```
