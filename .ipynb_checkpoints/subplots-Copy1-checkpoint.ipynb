```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Figure - canvas
#Axes - subplots
```

```python
figure,axes = plt.subplots(2,3)
x1 = np.array([2021,2022,2023,2024,2025,2026])
y1 = np.array([0,10000,60000,100000,200000,300000])
x2 = np.array(["jan","feb","mar","apr","may","june"])
y2 = np.array([5000,1000,3000,4500,500,7000])
x3 = np.array([12,10,11,14,7,6,10,14])
y3 = np.array([5000,4000,3000,7000,6000,2000,5000,6800])
x4 = np.array([10,12,14,9,12,13,5,2,21])
x5 = np.array(["Long Form","Shorts","Live","Posts"])
y5 = np.array([4060,1200,500,300])
x6 = np.array([11,12,10,8,7,6,9,16])
y6 = np.array([5500,3000,4000,7500,6500,1000,3000,9800])
axes[0,0].plot(x1,y1)
axes[0,1].bar(x2,y2, color = "green")
axes[0,2].pie(y5,labels = x5)
axes[1,0].hist(x4)
axes[1,1].scatter(x3,y3, color = "red")
axes[1,1].scatter(x6,y6)
plt.tight_layout()
#axes[1,2].grid(x1,y1)

plt.show()
```
