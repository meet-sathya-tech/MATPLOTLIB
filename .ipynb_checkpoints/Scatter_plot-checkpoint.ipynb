```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12,10,11,14,7,6,10,14])
y = np.array([5000,4000,3000,7000,6000,2000,5000,6800])

x1 = np.array([11,12,10,8,7,6,9,16])
y1 = np.array([5500,3000,4000,7500,6500,1000,3000,9800])
```

```python
plt.scatter(x,y,
           color = "red",
           alpha = 0.5,
           edgecolor = "black",
           label = "isatac_yoo")

plt.scatter(x1,y1,
           color = "blue",
           label = "isat_ac")

plt.title("Scatter Plot")
plt.legend()
```
